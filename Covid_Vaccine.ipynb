# Total cases vs Total Death World wide
* In 2021 America has the highest percentage of death



```python
#Total cases vs Total Death
query = psql.sqldf("SELECT location , population, date, total_cases, total_deaths, (total_deaths/total_cases* 100) as death_percentage FROM vaccine ORDER BY death_percentage")
query.head(50)
```

# Total cases Vs Total death in Bangladesh

* From  2020-03-02 to 2020-05-03 Bangladesh saw the highest death rate.
* In 2021 Bangladesh death rates stable but still there is 1% of people dying from COVID*19. 

```python
#Likelyhood of dying if you have covid in bangladesh
query = psql.sqldf("SELECT location , population, date, total_cases, total_deaths as total_deaths, (total_deaths*1.0*100)/total_cases as death_percentage FROM vaccine WHERE location ='Bangladesh' ORDER BY 3")
query.head(50)
```

# Likelyhood of getting infected in Bangladesh

* From 2020-06-19 to 2020-06-19 and 2020-07-08 to  2021-01-12, Bangladesh infection rates were at the peak.
* As new variants keep coming in 2021, Bangladesh infection rate never becomes stable from infection.

```python
#Likelyhood of getting infected in Bangladesh
query = psql.sqldf("SELECT location , date, total_cases, (total_cases/new_cases_smoothed_per_million)*100 as infected_percentage FROM vaccine WHERE location ='Bangladesh' ORDER BY 3")
query['infected_percentage'] = query['infected_percentage'].round(decimals=3)
query.head(50)
```

# Higest infection rate VS Population Worldwide

```python
# Looking at country that has highest infection rate compared to population
query = psql.sqldf("SELECT location ,population, Max(total_cases) as Highest_Infection_Count, Max((total_cases*100)/population) as Percent_Population_Infected FROM vaccine GROUP BY location, population ORDER BY Percent_Population_Infected")
query.head(50)
```

# Countries with Highest Infection Rate compared to Population
* Bangladesh rank at number 146 out of 171 countries

```python
# Countries with Highest Infection Rate compared to Population
# Bangladesh 146
query = psql.sqldf("SELECT location , MAX(cast(Total_deaths as int)) as total_death_count FROM vaccine Where continent is not null GROUP BY location ORDER BY total_death_count")
query.head(50)
```

# Highest death Vs Continent
* Sad to see Asia has the higest death rate.

```python
#Showing contintents with the highest death count per population
query = psql.sqldf("SELECT continent , MAX(cast(Total_deaths as int)) as total_death_count FROM vaccine Where continent is not null GROUP BY continent ORDER BY total_death_count desc")
query
```

# Death count inside Asian continent

* Out of 49 countries, Bangladesh is in number 7.
* Though the first Covid-19 was found in China there death rate is in number 17.
* It seems the new variant's consequences India faces the most.

```python
#Showing Asian with the highest death count per population
# Bangladesh in number 7
query = psql.sqldf("SELECT location,  MAX(cast(Total_deaths as int)) as total_death_count FROM vaccine Where continent= 'Asia' GROUP BY location ORDER BY total_death_count desc")
query
```
